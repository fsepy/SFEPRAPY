```python
%matplotlib inline

import io
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import interpolate
from distribution_auto_fit import best_fit_distribution
import scipy.stats as st

sns.set_style('ticks')
```

```python
data_csv = pd.read_csv('samples_car_park_qfd_andersonbell.csv', header=0, dtype=float)
cdf = data_csv['cdf']
q = data_csv['qfd']

cdf_func = interpolate.interp1d(cdf, q, kind='linear')

n_sample = 10000
samples_y = np.random.uniform(size=n_sample)
samples = np.sort(cdf_func(samples_y))

y, x = np.histogram(samples, bins=100, density=True)
x = (x + np.roll(x, -1))[:-1] / 2.0

print("Statistical Data")
fig, (ax1, ax2) = plt.subplots(nrows=1,ncols=2,constrained_layout=True,figsize=(3.94*2, 2.76));
ax1.plot(x, y);
ax1.set_xlabel('Fire load [kW/sq.m]');
ax1.set_ylabel('PDF');
ax2.plot(samples, np.linspace(0,1,len(samples)));
ax2.set_xlabel('Fire load [kW/sq.m]');
ax2.set_ylabel('CDF');
```

```python
fig, ax = plt.subplots(figsize=(3.94*2, 2.76*2));
dist_name, params = best_fit_distribution(samples, ax=ax)
print("Best: {}, Params: {}".format(dist_name, str(params)))
```

```python
print("Fitted Data")
y = np.linspace(0,1,1000);
x = st.norm.ppf(y, *(251.90611757771737, 7.015945381175767));
fig3, ax3 = plt.subplots(figsize=(3.94*2, 2.76*2));
ax3.plot(x, y, label='Fitted');
ax3.plot(q, cdf, label='Sampled')
ax3.set_xlabel('Sampled values');
ax3.set_ylabel('CDF');
ax3.legend().set_visible(True)
```
